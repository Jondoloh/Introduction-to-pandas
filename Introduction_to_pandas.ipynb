# 1. Part 1.

## Q1: Reading files and printing first five lines.

```python
#Importing libraries
import numpy as np
import pandas as pd
import io
```

```python
#Loading the data
data1 = pd.read_csv('/content/drive/MyDrive/af_austral.csv',skipinitialspace=True)
data1.columns = data1.columns.str.strip()
data1[['Names','Capital']] = data1[['Names','Capital']].apply(lambda x: x.str.strip())
data1.head()

```

```python
data2 = pd.read_csv('/content/drive/MyDrive/af_central.csv',skipinitialspace=True)
data2.columns = data2.columns.str.strip()
data2[['Names','Capital']] = data2[['Names','Capital']].apply(lambda x: x.str.strip())
data2.head()
```

```python
data3 = pd.read_csv('/content/drive/MyDrive/af_east.csv',skipinitialspace=True)
data3.columns = data3.columns.str.strip()
data3[['Names','Capital']] = data3[['Names','Capital']].apply(lambda x: x.str.strip())
data3.head()
```

```python
data4 = pd.read_csv('/content/drive/MyDrive/af_north.csv',skipinitialspace=True)
data4.columns = data4.columns.str.strip()
data4[['Names','Capital']] = data4[['Names','Capital']].apply(lambda x: x.str.strip())
data4.head()
```

```python
data5 = pd.read_csv('/content/drive/MyDrive/af_west.csv',skipinitialspace=True)
data5.rename(columns={list(data5)[0]:'Names'},inplace = True)
data5.columns = data5.columns.str.strip()
data5[['Names','Capital']] = data5[['Names','Capital']].apply(lambda x: x.str.strip())
data5.head()
```

### Q2: Merge all the files and add a new column that defines the region where the country belongs to.

```python
#Merging the files
df = pd.concat([data1,data2,data3,data4,data5],keys=["Austral","Central","East","North","West"]).reset_index()

# #del df['level_1']
df.drop('level_1',inplace=True,axis=1) #unwanted column

# #Adding a new column _Regions
df.rename(columns={list(df)[0]:'Region'},inplace = True)

# #Re_indexing the column names
data_merged = df.reindex(columns=['Names','Area','Population','Density','Capital','Region'])
data_merged.head()
```

```python
#confirming the count per region
data_merged.Region.value_counts()
```

### Q3: Calculate and present in the same Dataframe: The mean of the population $()$ of each region, the standard deviation () and the calculation of the coefficient of variation Cv and conclude: $$C_v = \frac{\sigma}{\mu}$$

```python
#coefficient of variation
def Cv(data):
  cv = data.std()/data.mean()
  return cv
```

```python
data_merged.groupby(['Region']).Population.agg([np.mean,np.std,Cv])
```

```python
#data_merged.groupby(['Region']).Population.agg(Cv)
```

```python
#Alternatively
cv = data_merged.groupby(['Region']).Population.std()/data_merged.groupby(['Region']).Population.mean()
results = data_merged.groupby(['Region']).Population.agg([np.mean,np.std])
results['Cv'] = cv
results
```

## 2. Part 2.

### Q1: Obtain the minimum area of each region and complete with the corresponding country name on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda p: p.loc[p.Area.idxmin()].loc[['Names','Area']])
```

### Q3: Get the minimum population of each region and complete with the corresponding country name on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda x:x.loc[x.Population.idxmin()].loc[['Names','Population']])
```

### Q4: Get the maximum population of each region and complete with the corresponding country name on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda x:x.loc[x.Population.idxmax()].loc[['Names','Population']])
```

### Q5: Obtain the average population of each region and complete with the name of country having the population closest to the average of the corresponding group, on the same line in the Serie or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda p:p.loc[(np.subtract(p.Population,p.Population.mean())).abs().idxmin()].loc[['Names','Population']])
```

### Q6: Obtain the average population of each region and complete with the name of the country having the population furthest from the average of the corresponding group, on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda p:p.loc[(np.subtract(p.Population,p.Population.mean())).abs().idxmax()].loc[['Names','Population']])
```

### Q7: Obtain the average density of each region and complete with the country name having the density closest to the average of the corresponding group, on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda p:p.loc[(np.subtract(p.Density,p.Density.mean())).abs().idxmin()].loc[['Names','Density']])
```

### Q8: Obtain the average density of each region and complete with the name of the country having the density furthest from the average of the corresponding group, on the same line in the Series or Dataframe obtained.

```python
data_merged.groupby(['Region']).apply(lambda p:p.loc[(np.subtract(p.Density,p.Density.mean())).abs().idxmax()].loc[['Names','Density']])
```

### Q9: Using the population and area fields, calculate the density with the apply and lambda functions.

```python
data_merged["Dens_Pop_Area"]=data_merged.apply(lambda p:(p.Population/p.Area),axis=1)
```

### Q10: Create a field that is a 10% increase in population using the apply and lambda functions.

```python
data_merged["10_Increase"]=data_merged.apply(lambda p:(p.Population*(1.1)),axis=1)
```
